# Creating movie recommendation system

#### Q. what is recommendation system ? 
##### Ans. suppose first time i visited to amazon, and i searched for iphone 14. now next time when i logged in amazon, it automatically
##### shows us products like iphone 13, iphone 13 pro max, iphone 12 etc.
##### So amazons shows us the recommendations based on our past orders/search/ behaviour/rating/tags (which we are searching on amazon).
##### It shows us the products which might intrest us. This is called recommendations system.

#### Types of recommendation systems
- Content based (tags)
- collaborative (rating, behaviour) : based on review by user 1, user 2 will also be recommended something. here multiple users are involved

We will discuss here only on Content based.
Let's say we are watching action movie in netflix. so based on this tag, netflix will recommend us other action movies to watch.
suppose i watched ironman, so it will recommend us avenger movie as well. because both are action movies and also contents are same.

## Feature selection : Seelct few columns which will help in createing Maching learning model

```python
movies.columns
```

```python
movies = movies[['id', 'title', 'genre', 'overview']] # select only those columns which are required 
movies.head(2)
```

```python
movies['tags'] = movies['overview'] + movies['genre']  # combine overview and genre column
movies.head(2)
```

```python
# drop the overview and genre column
movies_new_data = movies.drop(columns=['overview', 'genre'])
movies_new_data.head(2)
```

###### now we need to convert text data into vector using machine learning algorithm.
###### in NLP (Natural language processing,there are various methods to convert textdata into vector
- Bag of words
- TFIDF vectorizer

###### To use maching learning algorithm, we use scikit learn module https://pypi.org/project/scikit-learn/ 

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector = cv.fit_transform(movies_new_data['tags'].values.astype('U')).toarray() # converting tags column into UTF and then to vector
```

```python
vector
```

```python
vector.shape # rows and column
```

#### Cosine similarity
We will now use here cosine similarity, means how similar is ironman movie to the avengers.
it will find the similarity between 2 movies by finding the angle (theta) between 2 movies in dimensions.
it will find the top 5 movies which has smallest angle with the movie watched by users and show those 5 movies as recommendations

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

- Here we are checking for "The Godfather"which is index 2

```python
similarity[2]
```

```python
movies_new_data.head(5)
```

```python
movies_new_data[movies_new_data['title'] == 'The Godfather'].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse = True, key = lambda vector:vector[1])
for i in distance[0:5]:
    print(i)
    print(movies_new_data.iloc[i[0]].title)
    print('-'*5)
```

```python
def recommend(movie):
    index = movies_new_data[movies_new_data['title'] == movie].index[0] # find the index in dataframe
    distance = sorted(list(enumerate(similarity[index])), reverse = True, key = lambda vector:vector[1])
    for i in distance[0:5]:
        print(movies_new_data.iloc[i[0]].title)

recommend('Iron Man')
```
